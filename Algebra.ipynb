```python
import numpy as np
import matplotlib.pyplot as plt
A = np.array([[2, -1],[-1, 2]])
B = np.array([0, 3])
X = np.linalg.solve(A,B)
print(X) #Solution printing

plt.plot([0,A[0][0]],[0,A[1][0]], 'green', label = 'VA1 - first column vector of A')
plt.plot([0,A[0][1]],[0,A[1][1]], 'blue', label = 'VA2 - second column vector of A')

#Finding Scaled vectors
VA1_x_new = [0, A[0][0] * X[0]]
VA1_y_new = [0, A[1][0] * X[0]]
VA2_x_new = [0, A[0][1] * X[1]]
VA2_y_new = [0, A[1][1] * X[1]]
VA2_shifted_x = [VA1_x_new[1],VA2_x_new[1] + VA1_x_new[1]]
VA2_shifted_y = [VA1_y_new[1],VA2_y_new[1] + VA1_y_new[1]]

#linear combination of vectors
Solution_x = [0, VA2_shifted_x[1]]
Solution_y = [0, VA2_shifted_y[1]]
#plotting proccess
plt.plot(VA1_x_new, VA1_y_new, 'red', label = 'VA1_new-first  vector')
plt.plot(VA2_shifted_x, VA2_shifted_y, 'yellow', label = 'VA2_new-second vector')
plt.plot(Solution_x, Solution_y, 'black', label = 'linear combination of vectors')
plt.legend(prop={"size":8})
plt.grid()
```
